#### Based on above results, we have chosen this hypothesis statement 

A party needs 50% seats to form the government. It seems that INC has won most no. of voters but despite having a \
large no. of voters, they are not eligible to form the party as number of seats is less. To form the majority, \
440 candidates from INC took part in the elections of 8070 candidates. It is found that INC party has got 206 seats \
with sigma standard deviation 6.76. Total no. of seats that can be won are 543. \
What is inference obtained from this? 

50 % >= 272+ seats \
H0: p >= 0.5 \
H1: p < 0.5 \
n = 543 \
p_bar = 206/543 \
p0 = 0.5 \
Since sigma is known, we apply z score \
Since H1: p < 0.5, it is one side(left tailed) test. 

```python
# Z=(p_bar-po)/(sigma/squareroot(n))

z = ((206/543)-0.5)/(6.76/np.sqrt(543))
z
```

```python
import scipy.special as scsp
```

```python
# p_value = 0.5*(1+scsp.erf(z/np.sqrt(2))) 
p_value=0.5*(1+scsp.erf(-0.415/np.sqrt(2)))
p_value
```

Since p value is less than 0.5, the null hypothesis is rejected 

H1: Party with most no. of votes does not secure most no. of seats

We can see that INC with most no. of votes has not secured the most number of seats

#### The NULL Hypothesis is rejected
### Party with most no. of votes does not secure most no. of seats

Therefore we conclude that, if a party wants to form the government, it is dependent on total no. of seats a party secures and not on the no. of maximum voters. \
From the above results, we can predict that the party has a good chance of getting more votes in future, but total no. of seats is unsure.
